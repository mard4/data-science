major decrease in performance? <br>
Evaluate techniques beyond Decision Trees and Regression.

# Data Preprocessing

### Missing Values

```python
def checkNaNs(DF):
    """ 
    show only columns with missing values
    """
    print("Total rows in DF:", len(DF))
    print("Total NaNs in DF:", DF.isna().sum().sum())
    all_nans = DF.isna().sum()

    all_nans = all_nans[all_nans > 0].to_frame(name='Missing_Values')
    
    return all_nans
all_nans = checkNaNs(data)
print(all_nans)
import matplotlib.pyplot as plt
import seaborn as sns

# --- 1. Heatmap of Missing Values ---
plt.figure(figsize=(12, 6))
sns.heatmap(data[all_nans.index].isna(), cbar=False, cmap="viridis")
plt.title("Missing Values Heatmap")
plt.show()
```

```python
def remove_columns_nan(DF):
        """  
        Remove the columns where missing value are greater than 10% of total data
        """
        
        print("Columns in current dataset: ", len(DF.columns))    
        columns_to_remove = []
        for col,value in enumerate(pd.isna(DF).sum()):
            if (value/len(data))*100 > 10:
                #print(DF.columns[col])
                columns_to_remove.append(DF.columns[col])
        print("Columns with more than 10% of missing values: ",columns_to_remove)                       
        
        DF= DF.drop(columns=columns_to_remove)
        print("Columns in dataset after removal: ", len(DF.columns))       
        return DF
```

```python
def groupbypatients_checkna(DF):
    """
    Remove patients with more than 20% of features missing values (5/6 columns)
    """
    print("Size of dataset before removal:", len(DF))

    #print("Features/",print(6/(37-4) *100)
    # Count NaNs per row (patient-wise)
    DF['Num_NaNs'] = DF.isna().sum(axis=1)

    # Determine threshold dynamically (20% of total columns)
    threshold = int(0.2 * (DF.shape[1] - 1))  # Exclude 'Patient_ID' column from count
    index_patients_nans = DF.loc[DF['Num_NaNs'] >= threshold, 'Patient_ID']

    print(f"Num of patients with NaN >= {threshold}: {len(index_patients_nans)}")

    # Remove patients that exceed the threshold
    DF = DF[~DF['Patient_ID'].isin(index_patients_nans)].drop(columns=['Num_NaNs'])

    print("Size of dataset after removal:", len(DF))
    
    return DF
```

```python
def pre_processing(DF):
    """
    MAIN FUNCTION FOR PRE PROCESSING
    """
    data_wo_cols = remove_columns_nan(DF)
    DF = groupbypatients_checkna(data_wo_cols)
    return DF

df = pre_processing(data)
checkNaNs(df)
```

# Data Imputation & Data Mapping

```python
# Fever = Mean
# duration of pain ???? 
# extr nervous = mean

#  Imputation: K-Nearest Neighbors (KNN) Imputation, Regression Imputation

# per ora ho messo la Moda
df[all_nans.index] = df[all_nans.index].fillna(df[all_nans.index].mean())
```

```python
df.columns = df.columns.str.lower()

## removed columns:   ['Workoverload', 'Relationship_with_colleagues', 'Irrational_thoughts_work', 'Trauma', 'Incoordination', 'working_ability']

binary_cols = [
    'fever', 'sick_leave', 'earlier_hospitalization', 'familiy_history',
    'depression', 'stress', 'uses_analgesics', 'uses_corticosteroids', 'serious_disease',
    'neurogenic_signals', 'continuous_pain', 'nocturnal_pain', 'loss_muscle_strength',
    'failure_symptoms', 'paidwork'
]
    
df[binary_cols] = df[binary_cols].astype('bool')
```

```python
ordinal_mappings = {
    'duration_of_pain': {
        1: '0-1 weeks', 2: '1-2 weeks', 3: '3-4 weeks', 4: '4-5 weeks',
        5: '6-8 weeks', 6: '9-11 weeks', 7: '3-6 months', 8: '6-9 months',
        9: '9-12 months', 10: 'more than 1 year'
    },
    'weightloss_per_year': {
        1: 'decrease', 2: 'increase', 3: 'stable', 4: 'fluctuating'
    },
    'treatment': {
        1: 'advice', 2: 'rehabilitation', 3: 'surgery', 
        4: 'injection/medication', 5: 'combination of 1-4'
    }
}

for col, mapping in ordinal_mappings.items():
    df[col] = df[col].map(mapping).astype('category')
```

```python
numeric_cols = [
    'extremely_nervous', 'irrational_thoughts_risk_lasting',
    'coping_strategy', 'kinesiophobia_physical_exercise',
    'kinesiophobia_pain_stop', 'neck_pain_intensity', 'low_back_pain_intensity',
    'arm_left_pain_intensity', 'arm_right_pain_intensity', 'leg_left_pain_intensity',
    'leg_right_pain_intensity', 'decreased_mobility'
]
df[numeric_cols] = df[numeric_cols].astype(float)

df['age'] = df['age'].astype('category')
print(df.dtypes)
```
