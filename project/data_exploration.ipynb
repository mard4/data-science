## The dataset

-  1547 real patient cases on low back pain
- Outcome variable: Treatment 1-5

#### Research question
1. To what extend can one, based on the answers given in the questionnaire, automatically determine the
advised treatment for a patient with back pain.
2. To what extend can one reduce the number of questions, i.e. which questions can be left out without
major decrease in performance? <br>
Evaluate techniques beyond Decision Trees and Regression.

# Data Preprocessing

### Missing Values

```python
def checkNaNs(DF):
    print("Total rows in DF: ", len(DF))
    print("Total NaNs in DF: ",pd.isna(DF).sum().sum())
    #print(pd.isna(DF).sum())
    all_nans = pd.DataFrame(pd.isna(DF).sum())
    return all_nans
checkNaNs(data)
```

```python
def remove_columns_nan(DF):
        """  
        Remove the columns where missing value are greater than 10% of total data
        """
        
        print("Columns in current dataset: ", len(DF.columns))    
        columns_to_remove = []
        for col,value in enumerate(pd.isna(DF).sum()):
            if (value/len(data))*100 > 10:
                #print(DF.columns[col])
                columns_to_remove.append(DF.columns[col])
        print("Columns with more than 10% of missing values: ",columns_to_remove)                       
        
        DF= DF.drop(columns=columns_to_remove)
        print("Columns in dataset after removal: ", len(DF.columns))       
        return DF
        
data_wo_cols = remove_columns_nan(data)
```

```python
def groupbypatients_checkna(DF):
    """
    Remove patients with more than 20% of features missing values (5/6 columns)
    """
    print("Size of dataset before removal:", len(DF))

    #print("Features/",print(6/(37-4) *100)
    # Count NaNs per row (patient-wise)
    DF['Num_NaNs'] = DF.isna().sum(axis=1)

    # Determine threshold dynamically (20% of total columns)
    threshold = int(0.2 * (DF.shape[1] - 1))  # Exclude 'Patient_ID' column from count
    index_patients_nans = DF.loc[DF['Num_NaNs'] >= threshold, 'Patient_ID']

    print(f"Num of patients with NaN >= {threshold}: {len(index_patients_nans)}")

    # Remove patients that exceed the threshold
    DF = DF[~DF['Patient_ID'].isin(index_patients_nans)].drop(columns=['Num_NaNs'])

    print("Size of dataset after removal:", len(DF))
    
    return DF

data_wo_patients = groupbypatients_checkna(data_wo_cols)
```
