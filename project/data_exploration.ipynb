## The dataset

-  1547 real patient cases on low back pain
- Outcome variable: Treatment 1-5

#### Research question
1. To what extend can one, based on the answers given in the questionnaire, automatically determine the
advised treatment for a patient with back pain.
2. To what extend can one reduce the number of questions, i.e. which questions can be left out without
major decrease in performance? <br>
Evaluate techniques beyond Decision Trees and Regression.

# Data Preprocessing

### Missing Values

```python
def checkNaNs(DF):
    """ 
    show only columns with missing values
    """
    print("Total rows in DF:", len(DF))
    print("Total NaNs in DF:", DF.isna().sum().sum())
    all_nans = DF.isna().sum()

    all_nans = all_nans[all_nans > 0].to_frame(name='Missing_Values')
    
    return all_nans

def plotNaNs(DF):
    """ 
    Plot missing values heatmap
    """
    plt.figure(figsize=(12, 6))
    sns.heatmap(DF[checkNaNs(DF).index].isna(), cbar=False, cmap="viridis")
    plt.title("Missing Values Heatmap")
    plt.show()

all_nans = checkNaNs(data)
print(all_nans)
plotNaNs(data)
```

```python
def remove_columns_nan(DF):
        """  
        Remove the columns where missing value are greater than 10% of total data
        """
        
        print("Columns in current dataset: ", len(DF.columns))    
        columns_to_remove = []
        for col,value in enumerate(pd.isna(DF).sum()):
            if (value/len(data))*100 > 10:
                #print(DF.columns[col])
                columns_to_remove.append(DF.columns[col])
        print("Columns with more than 10% of missing values: ",columns_to_remove)                       
        
        DF= DF.drop(columns=columns_to_remove)
        print("Columns in dataset after removal: ", len(DF.columns))       
        return DF
```

```python
def groupbypatients_checkna(DF):
    """
    Remove patients with more than 20% of features missing values (5/6 columns)
    """
    print("Size of dataset before removal:", len(DF))

    #print("Features/",print(6/(37-4) *100)
    # Count NaNs per row (patient-wise)
    x_cols = [x for x in DF.columns if x != 'Patient_ID']
    DF['Num_NaNs'] = DF[x_cols].isna().sum(axis=1)

    # Determine threshold dynamically (20% of total columns)
    threshold = int(0.2 * (DF.shape[1] - 1))  # Exclude 'Patient_ID' column from count
    index_patients_nans = DF.loc[DF['Num_NaNs'] >= threshold, 'Patient_ID']

    print(f"Num of patients with NaN >= {threshold}: {len(index_patients_nans)}")

    # Remove patients that exceed the threshold
    DF = DF[~DF['Patient_ID'].isin(index_patients_nans)].drop(columns=['Num_NaNs'])

    print("Size of dataset after removal:", len(DF))
    
    return DF
```

```python
def pre_processing(DF):
    """
    MAIN FUNCTION FOR PRE PROCESSING
    """
    data_wo_cols = remove_columns_nan(DF)
    DF = groupbypatients_checkna(data_wo_cols)
    return DF

df = pre_processing(data)
checkNaNs(df)
```

# Create Databases

```python
# Convert entire DataFrame to object dtype if you have mixed types
df = df.astype(object)

# Replace NaNs with None (PostgreSQL interprets None as NULL)
df = df.where(pd.notnull(df), None)
df.fever.value_counts(dropna=False) 
```

```python
# create columns
df_new = df.copy()
dict_mapping = {'1': 'Advice', '2': 'Rehabilitation', '3': 'Surgery', '4': 'Injection/Medication', '5': 'Combination_of_1_4'}
df_new['treatment_description'] = df_new.apply(lambda x: dict_mapping[str(x['treatment'])], axis=1) 
df_new.rename(columns={'familiy_history': 'family_history'}, inplace=True)
```

```python
df_new.columns

# conversione
def convert_toBoolean(DF, boolean_columns):
    for col in boolean_columns:
        DF[col] = DF[col].map({'yes': True, 'no': False, '1': True, '0': False, True: True, False: False, np.nan: np.nan})
        DF[col] = DF[col].astype('boolean')
    return DF

def convert_toInt(DF, integer_columns):
    for col in integer_columns:
        DF[col] = pd.to_numeric(DF[col], errors='coerce')
        DF[col] = DF[col].astype('Int32')
    return DF

boolean_columns = [
    'family_history',
    'serious_disease',
    'fever',
    'uses_analgesics',
    'uses_corticosteroids',
    'neurogenic_signals',
    'loss_muscle_strength',
    'failure_symptoms',
    'depression',
    'stress',
    'sick_leave',
    'earlier_hospitalization',
    'paidwork'
]

integer_columns =  [
    'decreased_mobility',
    'weightloss_per_year',
    'extremely_nervous',
    'irrational_thoughts_risk_lasting',
    'coping_strategy',
    'kinesiophobia_physical_exercise',
    'kinesiophobia_pain_stop',
    'nocturnal_pain',
    'continuous_pain',
    'duration_of_pain',
    'neck_pain_intensity',
    'low_back_pain_intensity',
    'arm_left_pain_intensity',
    'arm_right_pain_intensity',
    'leg_left_pain_intensity',
    'leg_right_pain_intensity'
]


df_new = convert_toBoolean(df_new, boolean_columns)
df_new = convert_toInt(df_new, integer_columns)
df_new.dtypes
```

```python
from connection import * 
import numpy as np
from sqlalchemy import create_engine 
import psycopg2

DB_HOST = DB_HOST
DB_NAME = DB_NAME
DB_USER = DB_USER       
DB_PASS = DB_PASS       
DB_PORT = DB_PORT               # Default PostgreSQL port

CREATE_tables_sql = """
                    CREATE TABLE IF NOT EXISTS project.patient (
                   
                        patient_id SERIAL PRIMARY KEY,
                        age VARCHAR(50),
                        family_history BOOLEAN,
                        serious_disease BOOLEAN
                    );
                    
                    CREATE TABLE IF NOT EXISTS project.radvice (
                            treatment SERIAL PRIMARY KEY,
                            treatment_description VARCHAR(50),
                            patient_id INTEGER REFERENCES project.patient(patient_id)
                    );

                    CREATE TABLE IF NOT EXISTS project.clinical (
                        patient_id INTEGER REFERENCES project.patient(patient_id),
                        fever BOOLEAN,
                        uses_analgesics BOOLEAN,
                        uses_corticosteroids BOOLEAN,
                        neurogenic_signals BOOLEAN,
                        decreased_mobility INTEGER,
                        weightloss_per_year INTEGER,
                        loss_muscle_strength BOOLEAN,
                        failure_symptoms BOOLEAN
                    );
                    
                    CREATE TABLE IF NOT EXISTS project.psychological (
                        patient_id INTEGER REFERENCES project.patient(patient_id),
                        depression BOOLEAN,
                        extremely_nervous INTEGER,
                        stress BOOLEAN,
                        irrational_thoughts_risk_lasting INTEGER,
                        coping_strategy INTEGER,
                        kinesiophobia_physical_exercise INTEGER
                    );
                    
                    CREATE TABLE IF NOT EXISTS project.work (
                        patient_id INTEGER REFERENCES project.patient(patient_id),
                        sick_leave BOOLEAN,
                        earlier_hospitalization BOOLEAN,
                        paidwork BOOLEAN
                    );
                    
                    CREATE TABLE IF NOT EXISTS project.pain (
                        patient_id INTEGER REFERENCES project.patient(patient_id),
                        kinesiophobia_pain_stop INTEGER,
                        nocturnal_pain INTEGER,
                        continuous_pain INTEGER,
                        duration_of_pain INTEGER,
                        neck_pain_intensity INTEGER,
                        low_back_pain_intensity INTEGER,
                        arm_left_pain_intensity INTEGER,
                        arm_right_pain_intensity INTEGER,
                        leg_left_pain_intensity INTEGER,
                        leg_right_pain_intensity INTEGER
                    );
                    
       """
```

```python
##### Connecting
try:
    # Connect to PostgreSQL
    conn = psycopg2.connect(
        dbname=DB_NAME, user=DB_USER, password=DB_PASS, host=DB_HOST, port=DB_PORT
    )
    conn.autocommit = True  # Ensure changes are saved automatically

    # Create a cursor object
    cursor = conn.cursor()

    # Execute table creation queries
    cursor.execute(CREATE_tables_sql)

    print("Tables created successfully.")

    # Close the cursor and connection
    cursor.close()
    conn.close()

except Exception as e:
    print("Error:", e)
    
engine = create_engine(f'postgresql://{DB_NAME}:{DB_PASS}@{DB_HOST}:{DB_PORT}/{DB_NAME}')

# Create the tables in the database

# patient
df_new[['patient_id', 'age', 'family_history', 'serious_disease']]\
    .to_sql('patient', engine, schema='project', index=False, if_exists='replace')

# clinical
df_new[['patient_id', 'fever', 'uses_analgesics', 'uses_corticosteroids',
        'neurogenic_signals', 'decreased_mobility', 'weightloss_per_year',
        'loss_muscle_strength', 'failure_symptoms']]\
    .to_sql('clinical', engine, schema='project', index=False, if_exists='replace')

# psychological
df_new[['patient_id', 'depression', 'extremely_nervous', 'stress',
        'irrational_thoughts_risk_lasting', 'coping_strategy',
        'kinesiophobia_physical_exercise']]\
    .to_sql('psychological', engine, schema='project', index=False, if_exists='replace')

# work
df_new[['patient_id', 'sick_leave', 'earlier_hospitalization', 'paidwork']]\
    .to_sql('work', engine, schema='project', index=False, if_exists='replace')

# pain
df_new[['patient_id', 'kinesiophobia_pain_stop', 'nocturnal_pain', 'continuous_pain',
        'duration_of_pain', 'neck_pain_intensity', 'low_back_pain_intensity',
        'arm_left_pain_intensity', 'arm_right_pain_intensity',
        'leg_left_pain_intensity', 'leg_right_pain_intensity']]\
    .to_sql('pain', engine, schema='project', index=False, if_exists='replace')

# radvice
df_new[['treatment','treatment_description', 'patient_id']]\
    .to_sql('radvice', engine, schema='project', index=False, if_exists='replace')
```
